1. What is Simple Linear Regression (SLR)? Explain its purpose.

   Simple Linear Regression is a statistical method used to find the relationship between two variables — one independent variable (X) and one dependent variable (Y) — by fitting a straight line through the data points.

   Prediction:To predict the value of one variable (Y) based on the value of another variable (X).

      Understanding Relationship:To understand how one variable affects another — whether the relationship is positive (Y increases as X increases) or negative (Y decreases as X increases).
      
      Trend Analysis:To identify patterns or trends in data


2.  What are the key assumptions of Simple Linear Regression?

      Linearity: The relationship between X (input) and Y (output) should be a straight line.If X increases, Y should increase or decrease in a straight pattern — not curved.

      Independence of Errors: Each data point should be independent of the others.The errors (differences between actual and predicted values) should not depend on each other.

      Equal Variance (Homoscedasticity):The spread of errors should be almost the same across all values of X.The errors should not get bigger or smaller as X changes.

      Normal Distribution of Errors: The errors (residuals) should follow a normal (bell-shaped) curve.This helps the model make reliable predictions and tests.
      
      No Autocorrelation (for time data):In time-based data, the errors should not follow any pattern over time.

3. Write the mathematical equation for a simple linear regression model and
explain each term.

      Y=a+bX+e

      Y (Dependent Variable):The variable we want to predict or explain.


      X (Independent Variable):The variable used to predict Y.



      a (Intercept or Constant):The value of Y when X = 0.It shows where the regression line cuts the Y-axis.

      b (Slope or Coefficient):It shows how much Y changes when X increases by 1 unit.



      e (Error term or Residual):The difference between the actual value and the predicted value.It represents random factors not explained by the model.

4. Provide a real-world example where simple linear regression can be
applied.

      A real-world example of Simple Linear Regression is predicting a student’s exam marks based on the number of hours studied.

In this case:

Dependent variable (Y): Exam marks

Independent variable (X): Hours studied

The regression equation can be written as:

Y=a+bX

where Y is the predicted marks, a is the intercept, and b is the slope that shows how much the marks increase for each additional hour of study.

5. What is the method of least squares in linear regression?

      The Method of Least Squares is a mathematical technique used in linear regression to find the best-fitting line through a set of data points.

It works by minimizing the sum of the squared differences between the actual values (Y) and the predicted values (Ŷ) given by the regression line.

7. Name and briefly describe three common evaluation metrics for regression
models.

      Mean Absolute Error (MAE):It measures the average of the absolute differences between actual and predicted values.It shows how much, on average, the predictions are off.Simple meaning: Lower MAE = better model.

      Mean Squared Error (MSE):It measures the average of squared errors between actual and predicted values.Squaring gives more weight to large errors.Simple meaning: Smaller MSE = more accurate predictions.

      R-squared:It shows how well the regression model explains the variation in the dependent variable.Value lies between 0 and 1 — closer to 1 means a better fit.Simple meaning: R² tells how much of the change in Y is explained by X.

8. What is the purpose of the R-squared metric in regression analysis?

      The R-squared  metric measures how well a regression model explains the variation in the dependent variable (Y) using the independent variable(s) (X).It shows the goodness of fit of the model — in simple words, how well the regression line fits the data points.


9. Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.
(Include your Python code and output in the code box below.)


```python
from sklearn.linear_model import LinearRegression
import numpy as np


X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([3, 4, 2, 5, 6])


model = LinearRegression()
model.fit(X, y)


print("Slope (b):", model.coef_[0])
print("Intercept (a):", model.intercept_)
```
